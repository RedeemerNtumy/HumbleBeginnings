Matplotlib oo method used for more advanced visualisations
pyplot for quick visualisations

```python
over_50=heart_disease[heart_disease.age>50]
over_50
```

```python
#pyplot
over_50.plot.scatter(x="age",y="chol",c="target");
```

```python
#oo method with pyplot
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot.scatter(x="age",y="chol",c="target",ax=ax);
```

```python
#Real OO
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax.set(title="Heart Disease and cholestrol levels",xlabel="Age",ylabel="Cholestrol level")
ax.axhline(over_50["chol"].mean(),linestyle="--");
ax.legend(*scatter.legend_elements(),title="Target");
```

```python
over_50
```

Subplot of chol,age,thalach

```python
# Ax0 which reprsents the first subplot on the figure
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))
scatter1=ax0.scatter(x=over_50.age,y=over_50.chol,c=over_50.target)
ax0.set(title="Heart Disease and Cholestrol levels",xlabel="Age",ylabel="Cholestrol");
ax0.legend(*scatter1.legend_elements(),title="Target")
ax0.axhline(over_50.chol.mean(),linestyle="--",color="green");

#Ax1 for the thalach
scatter2=ax1.scatter(x=over_50.age,y=over_50.thalach,c=over_50.target)
ax1.set(title="Heart Disease and Heart Rate",xlabel="Age",ylabel="Heart Rate")
ax1.legend(*scatter1.legend_elements(),title="Target")
ax1.axhline(over_50.thalach.mean(),linestyle="--",color="blue");
```

```python
#Using sex as the legend instead of target
#The plot below show the heart disease sample results of individuals over 50 grouped by sex
fig,ax=plt.subplots(figsize=(10,6))
ax.set(title="Heart disease presentation of old men and women",xlabel="trestbps",ylabel="Age")
scatter=ax.scatter(x=over_50.age,y=over_50.trestbps,c=over_50.sex)
ax.legend(*scatter.legend_elements(),title="Sex")
ax.axhline(over_50.trestbps.mean(),linestyle="--",color="red");
male=[]
for a in over_50.sex:
    if a==1:
        male.append(a)
might_die=[]
for i in over_50.trestbps:
    if i >over_50.trestbps.mean():
        might_die.append(i)
print(f"From this, we find out that out of {len(over_50.age)} patiens, {len(might_die)} patients have a trestbps above normal. There are {len(male)} males and {len(over_50.sex)-len(male)} females")
```

Making them share the same x axis

```python
# Ax0 which reprsents the first subplot on the figure
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
scatter1=ax0.scatter(x=over_50.age,y=over_50.chol,c=over_50.target)
ax0.set(title="Heart Disease and Cholestrol levels",ylabel="Cholestrol");
ax0.legend(*scatter1.legend_elements(),title="Target")
ax0.axhline(over_50.chol.mean(),linestyle="--",color="green");

#Ax1 for the thalach
scatter2=ax1.scatter(x=over_50.age,y=over_50.thalach,c=over_50.target)
ax1.set(title="Heart Disease and Heart Rate",xlabel="Age",ylabel="Heart Rate")
ax1.legend(*scatter1.legend_elements(),title="Target")
ax1.axhline(over_50.thalach.mean(),linestyle="--",color="blue");
#Add Title Figure
fig.suptitle("Heart Disease Analysis",fontsize=18,fontweight="bold");
```

```python
plt.style.use("seaborn-whitegrid") #Produces black and white graph unless we use cmap
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")#cmap changes the color scheme
ax.set(title="Heart Disease and cholestrol levels",xlabel="Age",ylabel="Cholestrol level")
ax.axhline(over_50["chol"].mean(),linestyle="--",color="red");
ax.legend(*scatter.legend_elements(),title="Target");
```

```python
# Customizing x and y lims
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
scatter1=ax0.scatter(x=over_50.age,y=over_50.chol,c=over_50.target,cmap="winter")
ax0.set(title="Heart Disease and Cholestrol levels",ylabel="Cholestrol",xlim=50);
ax0.legend(*scatter1.legend_elements(),title="Target")
ax0.axhline(over_50.chol.mean(),linestyle="--",color="red");


scatter2=ax1.scatter(x=over_50.age,y=over_50.thalach,c=over_50.target,cmap="winter")
ax1.set(title="Heart Disease and Heart Rate",xlabel="Age",ylabel="Heart Rate",ylim=(60,200))
ax1.legend(*scatter1.legend_elements(),title="Target")
ax1.axhline(over_50.thalach.mean(),linestyle="--",color="red");
#Add Title Figure
fig.suptitle("Heart Disease Analysis",fontsize=18,fontweight="bold");
```
